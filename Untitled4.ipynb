```python
import matplotlib.pyplot as plt

# Sample data (replace with your chord frequencies)
chord_frequencies = {'C': 20, 'G': 15, 'D': 10, 'A': 5, 'E': 8}

# Create the histogram
plt.bar(chord_frequencies.keys(), chord_frequencies.values())  # Keys for x-axis, values for y-axis
plt.xlabel("Chord")
plt.ylabel("Frequency")
plt.title("Chord Frequency Distribution")
plt.show()
```
