```python
date = pd.date_range('20200101', periods=12, freq='M')
type(date)
```

```python
date
```

```python
date = pd.Series(date)
date
```

```python
type(date)
```

```python
date.dt.days_in_month
```

```python
death_toll["Days"]= date.dt.days_in_month
```

```python
death_toll
```

```python
death_toll.loc[11, "Days"] = 15
```

```python
death_toll
```

```python
death_toll["Total_deaths"] = death_toll["New_deaths"]*death_toll["Days"]
```

```python
death_toll
```

```python
death_toll["Total_deaths"].cumsum()
```

```python
death_toll["Cum_deaths"] = death_toll["Total_deaths"].cumsum()
```

```python
death_toll
```

```python
cum = death_toll["Cum_deaths"]
```

```python
cum
```

```python
march = cum[2]
june = cum[5]
october = cum[9]
december = cum[11]
```

```python
deaths = [march, june, october, december]
proportions = [round(death/december, 2) for death in deaths]

proportions 
```

```python
xmax_vals = [round(0.5 + proportion * 0.3, 3) for proportion in proportions]
```

```python
xmax_vals
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=4, ncols=1,figsize=(6,8))
axes = [ax1, ax2, ax3, ax4]

for ax in axes:
    ax.plot(death_toll['Month'], death_toll['New_deaths'],
            color='#af0b1e', alpha=0.1)
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.tick_params(bottom=0, left=0) 
    for location in ['left', 'right', 'top', 'bottom']:
        ax.spines[location].set_visible(False)

ax1.plot(death_toll['Month'][:3], death_toll['New_deaths'][:3],color='#EE83DC', linewidth=2.5)
ax2.plot(death_toll['Month'][2:6], death_toll['New_deaths'][2:6],color='#EE83DC', linewidth=2.5)
ax3.plot(death_toll['Month'][5:10], death_toll['New_deaths'][5:10],color='#EE83DC', linewidth=2.5)
ax4.plot(death_toll['Month'][9:12], death_toll['New_deaths'][9:12],color='#EE83DC', linewidth=2.5)

ax1.text(0.5, -80, '0', alpha=0.5)
ax1.text(3.5, 2000, '1,844', alpha=0.5)
ax1.text(11.5, 2400, '2,247', alpha=0.5)

ax1.text(1.1, -300, 'Jan - Mar', color='#af0b1e',weight='bold', rotation=3)
ax2.text(3.7, 800, 'Mar - Jun', color='#af0b1e', weight='bold')
ax3.text(7.1, 500, 'Jun - Oct', color='#af0b1e', weight='bold')
ax4.text(10.5, 600, 'Oct - Dec', color='#af0b1e', weight='bold',rotation=45)

ax1.text(0.5, 3500, 'The virus kills 851 people each day',size=14, weight='bold',color = '#9B5C29')
ax1.text(0.5, 3150, 'Average number of daily deaths per month in the US',size=12,color = '#9B5C29')

for ax, xmax, death in zip(axes, xmax_vals, deaths):
    ax.axhline(y=1600, xmin=0.5, xmax=0.8,linewidth=6, color='#BC8CF2',alpha=0.1)
    ax.axhline(y=1600, xmin=0.5, xmax=xmax,linewidth=6, color='#BC8CF2')
    ax.text(7.5, 1850, format(death, ','),color='#af0b1e', weight='bold')
plt.show()
```
