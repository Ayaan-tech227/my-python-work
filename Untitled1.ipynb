```python
import matplotlib.pyplot as lib
import pandas as pd
```

```python
data = pd.read_csv("WHO.csv")
```

```python
data
```

```python
data.info()
```

```python
data.dtypes
```

```python
data["Date_reported"] = pd.to_datetime(data["Date_reported"])
```

```python
data.info()
```

```python
data
```

```python
bo = data["Country"] == "China"
```

```python
bo
```

```python
bol = data[bo]
```

```python
bol
```

```python
lib.plot(bol["Date_reported"],bol["Cumulative_cases"])
lib.title("India : cumulative cases at every date")
lib.xlabel("dates")
lib.ylabel("number of patients")
lib.show()
```

```python
bos = data["Country"] == "Pakistan"
```

```python
bols = data[bos]
```

```python
bosll = data["Country"] == "Panama"
```

```python
bolsll = data[bosll]
```

```python
bolsll
```

```python
lib.plot(bolsll["Date_reported"],bolsll["Cumulative_cases"],label = "Panama")
lib.title("India and Pakistan : cumulative cases at every date")
lib.xlabel("Dates")
lib.ylabel("Number of patients")
lib.legend()
lib.show()
```
