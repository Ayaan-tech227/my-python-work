```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('data_science.csv')
```

```python
df
```

```python
df.drop(df[['Unnamed: 0','salary']],axis = 1,inplace = True)
```

```python
df1 = df.groupby('work_year')['salary_in_usd'].mean().round(2)
```

```python
df1
```

```python
data = {
    'work_year':df1.index,
    'avrage_salary':df1.values
}
```

```python
data_frame = pd.DataFrame(data)
```

```python
data_frame
```

```python
data_frame['avrage_salary'] = (data_frame['avrage_salary']/1000).round(2)
```

```python
data_frame
```

```python
data_frame.plot(kind = 'bar',x = 'work_year',y = 'avrage_salary',legend = True)
```

```python
ax = data_frame.plot(kind = 'bar',x = 'work_year',y = 'avrage_salary',legend = True)
ax.bar_label(ax.containers[0],labels = data_frame['avrage_salary'].map('{:.2f}k'.format))
plt.show()
```

```python
df5 = df[["salary_in_usd", "company_size"]]
df5
```

```python
S = df5[df5['company_size'] == 'S']
M = df5[df5['company_size'] == 'M']
L = df5[df5['company_size'] == "L"]

labels = ['Small','Medium', 'Large']
sal_mean = [S['salary_in_usd'].mean(), M["salary_in_usd"].mean(), L["salary_in_usd"].mean()]
sal_mean
```

```python
label_change = np.round([x/1000 for x in sal_mean], 2)
label_change = list(map(str, label_change))
label_change = [x + ' K' for x in label_change]
label_change
```

```python
z = plt.bar(labels, sal_mean)
plt.bar_label(z,label_change, alpha = 0.8)
plt.title('Distribution of salary by comapny size')
plt.xlabel('Size of company')
plt.ylabel('Salary in usd')
plt.show()
```

```python
sns.histplot(S['salary_in_usd'],label = 'Small',kde = True)
sns.histplot(M['salary_in_usd'],label = 'Small',kde = True)
sns.histplot(L['salary_in_usd'],label = 'Small',kde = True)
plt.title('Distribution of salary by comapny size')
plt.xlabel('Size of company')
plt.ylabel('Salary in usd')
plt.show()
```
