```python
date = pd.date_range('20200101',periods = 12,freq = 'M')
```

```python
date = pd.Series(date)
```

```python
d = date.dt.days_in_month
```

```python
dt['date'] = date.dt.days_in_month
```

```python
dt
```

```python
dt.loc[11, "date"] = 15
```

```python
dt
```

```python
dt['Total_deaths'] = dt['New_deaths']*dt['date']
```

```python
dt
```

```python
ds = dt['Total_deaths'].cumsum
```

```python
dt['cum_deaths'] = dt['Total_deaths'].cumsum()
```

```python
dt
```

```python
cum = dt['cum_deaths']
```

```python
march = cum[2]
june = cum[5]
october = cum[9]
december = cum[11]
```

```python
deaths = [march, june, october, december]
properties = [round(d/december, 2)for d in deaths]
```

```python
properties
```

```python
Xmax_Vals = [round(0.5 + propertie * 0.3,2) for propertie in properties]
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=4, ncols=1,figsize=(6,8))

axes = [ax1, ax2, ax3, ax4]
for i in axes:
    
    i.plot(dt['Month'],dt['New_deaths'], alpha=0.1,color='#0E2AB9')
    i.set_yticklabels([])
    i.set_xticklabels([])
    i.tick_params(bottom=0, left=0) 
    for location in ['left', 'right','top','bottom']:
        i.spines[location].set_visible(False)

ax1.plot(dt['Month'][:3], dt['New_deaths'][:3],color='#0E2AB9', linewidth=2.5)
ax2.plot(dt['Month'][2:6], dt['New_deaths'][2:6],color='#0E2AB9', linewidth=2.5)
ax3.plot(dt['Month'][5:10], dt['New_deaths'][5:10],color='#0E2AB9', linewidth=2.5)
ax4.plot(dt['Month'][9:12], dt['New_deaths'][9:12],color='#0E2AB9', linewidth=2.5)

ax1.text(1.1, -300, 'Jan - Mar', color='#DE7B18',weight='bold', rotation=3)
ax2.text(3.7, 800, 'Mar - Jun', color='#DE7B18', weight='bold')
ax3.text(7.1, 500, 'Jun - Oct', color='#DE7B18', weight='bold')
ax4.text(10.5, 600, 'Oct - Dec', color='#DE7B18', weight='bold',rotation=45)

ax1.text(0.5, -80, '0', alpha=0.5)
ax1.text(3.5, 2000, '1,844', alpha=0.5)
ax1.text(11.5, 2400, '2,247', alpha=0.5)
ax1.text(0.5, 3500, 'The virus kills 851 people each day',size=14, weight='bold',color = '#08165D')
ax1.text(0, 3150, 'Average number of daily deaths per month in the US',size=12,color = '#DE7B18')
for ax, xmax, death in zip(axes, Xmax_Vals, deaths):
    ax.axhline(y=1600, xmin=0.5, xmax=0.8,linewidth=6, color='#DE7B18',alpha=0.1)
    ax.axhline(y=1600, xmin=0.5, xmax=xmax,linewidth=6, color='#0E2AB9')
    ax.text(7.5, 1850, format(death, ','),color='#08165D', weight='bold')
plt.show()
```
