```python
html = pd.read_html("https://en.wikipedia.org/wiki/Economy_of_the_United_States")
html
```

```python
html[0]
```

```python
html[3]
```

```python
x = html[3].Year
```

```python
y = html[3]['Inflation rate (in Percent)']
y
```

```python
y= y.str.replace('%','').astype(float)
y
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y, marker='o', color='b', label='Blue Circles')  # Blue circles
plt.xlabel('Years')
plt.ylabel('Inflation rate (in Percent)')
plt.title('Inflation rate (in Percent) by year growth ')
plt.legend()

plt.show()
```
