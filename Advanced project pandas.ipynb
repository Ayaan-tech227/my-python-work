```python
countries = f500["country"].unique()
countries
```

```python
top_employer_by_country = {}

for c in countries:
    df = f500[f500["country"] == c]
    sort_employees = df.sort_values("employees", ascending=False)
    top_employer = sort_employees.iloc[0,0]
    top_employer_by_country[c] = top_employer
```

```python
top_employer_by_country
```

```python
top_employer_by_country = pd.Series()

for c in countries:
    df = f500[f500["country"] == c]
    sort_employees = df.sort_values("employees", ascending=False)
    top_employer = sort_employees.iloc[0,0]
    top_employer_by_country[c] = top_employer
```
